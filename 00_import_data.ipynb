```python
# Importing packages
import pandas as pd
import sqlite3
```

```python
# Reads file and defines row 4 as header
df =pd.read_csv("traffic.csv")
df.head() 
```

```python
#Initiate a new SQLite database connection object and assign this object to the variable 'dn_conn'
db_conn = sqlite3.connect("LF_Case.db")

# establish a cursor object and assign to c
c = db_conn.cursor()

# Create empty table for actual generation
c.execute(
    """
    CREATE TABLE IF NOT EXISTS traffic(
        event TEXT NOT NULL,
        date DATE NOT NULL,
        country TEXT,
        city TEXT,
        artist TEXT,
        album TEXT, 
        track TEXT,
        isrc TEXT, 
        linkid TEXT);
     """
)

# Populate Tables
df.to_sql('traffic', db_conn, if_exists='replace', index=False)

pd.read_sql("SELECT * FROM traffic LIMIT 10", db_conn)
```

## 1. How many total pageview events did the links in the provided dataset receive in the full period, how many per day?

```python
# Total number of pageviews
pd.read_sql("""
    SELECT 
        COUNT(*) as total_pageviews
    FROM traffic 
    WHERE event='pageview'"""
, db_conn)
```

```python
# Number of pageviews per day
pd.read_sql("""
    SELECT 
        date, 
        COUNT(*) as pageviews 
    FROM traffic 
    WHERE event='pageview'
    GROUP BY date"""
, db_conn)
```

## 2. What about the other recorded events?

```python
# Total number of clicks and previews
pd.read_sql("""
    SELECT 
        COUNT(*) as other_events 
    FROM traffic 
    WHERE event!='pageview'"""
, db_conn)
```

```python
# Number of clicks and previews per day
pd.read_sql("""
    SELECT 
        date, 
        COUNT(*) as other_events  
    FROM traffic 
    WHERE event!='pageview' 
    GROUP BY date"""
, db_conn)
```

```python
pd.read_sql("""
    SELECT 
        distinct(event) 
    FROM traffic 
    WHERE event!='pageview'"""
, db_conn)
```

## 3. Which countries did the pageviews come from?

```python
# Number of pageviews per country
pd.read_sql("""
    SELECT 
        country, 
        COUNT(*) as pageviews 
    FROM traffic 
    WHERE event='pageview' GROUP BY country"""
, db_conn)
```

## 4. What was the overall click rate (clicks/pageviews)?

```python
# Calculates click rates as number of clicks over number of pageviews 
pd.read_sql(
"""
SELECT 
    SUM(CASE WHEN event='pageview' THEN 1 ELSE 0 END) AS pageviews, 
    SUM(CASE WHEN event='click' THEN 1 ELSE 0 END) AS clicks, 
    CAST(SUM(CASE WHEN event='click' THEN 1 ELSE 0 END) AS FLOAT) / SUM(CASE WHEN event='pageview' THEN 1 ELSE 0 END) AS clickrate
FROM traffic;
"""
    , db_conn)

```

## 5. How does the clickrate distribute across different links?


```python
# Calculates click rates as number of clicks over number of pageviews
pd.read_sql(
"""
SELECT 
    linkid,
    artist,
    album, 
    track,
    SUM(CASE WHEN event='pageview' THEN 1 ELSE 0 END) AS pageviews, 
    SUM(CASE WHEN event='click' THEN 1 ELSE 0 END) AS clicks, 
    CAST(SUM(CASE WHEN event='click' THEN 1 ELSE 0 END) AS FLOAT) / SUM(CASE WHEN event='pageview' THEN 1 ELSE 0 END) AS clickrate
FROM traffic
GROUP BY linkid
ORDER BY pageviews DESC;
"""
    , db_conn)
```
